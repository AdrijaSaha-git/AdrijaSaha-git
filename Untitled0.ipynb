```python
# === IMPORTS & SETTINGS ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from google.colab import files
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 140)

# === UPLOAD FILE ===
print("Upload your file (ecommerce_orders_dataset.csv) using the dialog...")
uploaded = files.upload()
fname = next(iter(uploaded.keys()))
print("Loaded:", fname)

# === READ DATA ===
df = pd.read_csv(fname, low_memory=False)
print("\nInitial shape:", df.shape)

# Replace literal string 'nan' with real NaN
df = df.replace(["nan", "NaN", ""], np.nan)

# ---------------------------
# Confirm exact columns (for debugging)
print("\nColumns detected (first 80 chars):")
print(df.columns.tolist())

# ---------------------------
# 1) PREVIEW & SUMMARY
print("\n=== 1) PREVIEW & SUMMARY ===")
n_rows, n_cols = df.shape
print("Number of rows:", n_rows)
print("Number of columns:", n_cols)

missing_counts = df.isna().sum().sort_values(ascending=False)
print("\nMissing values per column (top 50):")
display(missing_counts.head(50))

print("\nData types summary:")
display(df.dtypes)

missing_counts.to_csv("missing_counts.csv")
print("Saved missing_counts.csv")

# ---------------------------
# 2) DATA CLEANING
print("\n=== 2) DATA CLEANING ===")
# Standardize payment_mode and region casing
for col in ["payment_mode", "region"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().replace({"nan": np.nan})
        def normalize_pm(x):
            if pd.isna(x):
                return x
            s = x.strip()
            # keep common acronyms uppercase
            acr = {"cod":"COD","upi":"UPI"}
            if s.lower() in acr:
                return acr[s.lower()]
            # common words title case
            return s.title()
        df[col] = df[col].apply(normalize_pm)

# Fill missing region with mode (most common region), fallback to "Unknown"
if "region" in df.columns:
    mode_region = df["region"].mode()
    fill_region = mode_region.iloc[0] if len(mode_region)>0 else "Unknown"
    df["region"] = df["region"].fillna(fill_region)
    df["region"] = df["region"].fillna("Unknown")
print("Filled missing 'region' with:", fill_region)

# ---------------------------
# 3) COLUMN REDUCTION
print("\n=== 3) COLUMN REDUCTION ===")
# Drop custom_col_1 ... custom_col_30 if present
custom_cols = [f"custom_col_{i}" for i in range(1,31)]
present_custom = [c for c in custom_cols if c in df.columns]
print("Found custom cols to drop:", present_custom)
df.drop(columns=present_custom, inplace=True, errors='ignore')
print("Shape after dropping custom cols:", df.shape)
pd.Series(df.columns.tolist(), name="remaining_columns").to_csv("remaining_columns.csv", index=False)
print("Saved remaining_columns.csv")

# ---------------------------
# Helpers: canonical column names given your header
order_col = "order_id"                 # as provided
cust_col = "customer_id"
prod_col = "product_id"
pc_col = "product_category"
unit_price_col = "unit_price"
qty_col = "quantity"
discount_col = "discount_percent"
payment_col = "payment_mode"
region_col = "region"
segment_col = "customer_segment"
date_col = "order_date"
delivery_days_col = "delivery_days"
delivery_status_col = "delivery_status"
prime_col = "is_prime_customer"
returned_col = "is_returned"
rating_col = "order_rating"
feedback_col = "customer_feedback"
delivery_notes_col = "delivery_notes"
delivery_date_col = "delivery_date"
total_value_col = "total_value"
return_amount_col = "return_amount"
profit_margin_col = "profit_margin"

# Ensure numeric conversions where needed
for c in [unit_price_col, qty_col, discount_col, delivery_days_col, total_value_col, return_amount_col, profit_margin_col, rating_col]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# ---------------------------
# 4) TOTAL REVENUE BY CATEGORY
print("\n=== 4) Top 5 Product Categories by Total Revenue ===")
if pc_col not in df.columns or total_value_col not in df.columns:
    raise KeyError("Missing expected columns for task 4. Ensure product_category and total_value exist.")
top5 = df.groupby(pc_col)[total_value_col].sum().sort_values(ascending=False).head(5).reset_index()
display(top5)
top5.to_csv("top5_categories_by_revenue.csv", index=False)

plt.figure(figsize=(8,5))
sns.barplot(data=top5, x=total_value_col, y=pc_col)
plt.title("Top 5 Product Categories by Total Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")
plt.tight_layout()
plt.savefig("top5_categories_by_revenue.png", dpi=150)
print("Saved top5_categories_by_revenue.png")

# ---------------------------
# 5) PRIME vs NON-PRIME
print("\n=== 5) Prime vs Non-Prime ===")
# Normalize prime_col to 'Prime'/'Non-Prime'
df[prime_col] = df[prime_col].astype(str).str.strip().replace({"nan": np.nan})
df[prime_col] = df[prime_col].fillna("Non-Prime")
df[prime_col] = df[prime_col].apply(lambda x: "Prime" if str(x).strip().lower() in ["1","true","yes","prime","y"] else "Non-Prime")

prime_stats = df.groupby(prime_col).agg(
    avg_order_rating = pd.NamedAgg(column=rating_col, aggfunc=lambda x: round(x.dropna().mean(),3)),
    avg_total_value = pd.NamedAgg(column=total_value_col, aggfunc=lambda x: round(x.dropna().mean(),2)),
    count_orders = pd.NamedAgg(column=order_col, aggfunc="count")
).reset_index()
display(prime_stats)
prime_stats.to_csv("prime_vs_nonprime_stats.csv", index=False)

# Simple inference
print("\nInference (auto):")
if set(["Prime","Non-Prime"]).issubset(set(prime_stats[prime_col].values)):
    p = prime_stats.set_index(prime_col)
    if p.loc["Prime","avg_total_value"] > p.loc["Non-Prime","avg_total_value"]:
        print("- Prime customers have higher average order value than Non-Prime.")
    else:
        print("- Non-Prime customers have higher or equal average order value than Prime.")
    if p.loc["Prime","avg_order_rating"] > p.loc["Non-Prime","avg_order_rating"]:
        print("- Prime customers have higher average rating than Non-Prime.")
    else:
        print("- Non-Prime customers have higher or equal rating than Prime.")
else:
    print("- One of Prime/Non-Prime groups has 0 rows; see prime_vs_nonprime_stats.csv")

# ---------------------------
# 6) DELIVERY TIME vs RATING
print("\n=== 6) Delivery Time vs Rating ===")
if delivery_days_col not in df.columns or rating_col not in df.columns:
    raise KeyError("Missing delivery_days or order_rating column.")
sub = df[[delivery_days_col, rating_col]].dropna()
if len(sub) < 2:
    print("Not enough data to compute correlation.")
else:
    corr = sub.corr().iloc[0,1]
    print(f"Pearson correlation (delivery_days vs order_rating): {corr:.4f}")
    plt.figure(figsize=(7,5))
    sns.regplot(data=sub, x=delivery_days_col, y=rating_col, scatter_kws={'s':10}, line_kws={'linewidth':1})
    plt.title(f"Delivery days vs Order rating (corr={corr:.3f})")
    plt.tight_layout()
    plt.savefig("delivery_vs_rating.png", dpi=150)
    print("Saved delivery_vs_rating.png")
    if abs(corr) < 0.1:
        print("Thought: very weak or no linear relationship between delivery time and rating.")
    elif corr < 0:
        print("Thought: weak-to-moderate negative relationship — longer delivery tends to reduce ratings.")
    else:
        print("Thought: positive relationship — investigate (could be confounding).")

# ---------------------------
# 7) RETURNS ANALYSIS
print("\n=== 7) Returns Analysis ===")
if returned_col not in df.columns:
    raise KeyError("Missing is_returned column.")
df[returned_col] = pd.to_numeric(df[returned_col], errors='coerce').fillna(0)
ret_by_cat = df.groupby(pc_col).agg(
    orders = pd.NamedAgg(column=order_col, aggfunc="count"),
    returned = pd.NamedAgg(column=returned_col, aggfunc="sum")
).reset_index()
ret_by_cat["return_rate_pct"] = (ret_by_cat["returned"] / ret_by_cat["orders"] * 100).round(2)
ret_by_cat = ret_by_cat.sort_values("return_rate_pct", ascending=False)
display(ret_by_cat.head(20))
ret_by_cat.to_csv("returns_by_category.csv", index=False)
if len(ret_by_cat)>0:
    highest = ret_by_cat.iloc[0]
    print(f"Highest return rate: {highest[pc_col]} = {highest['return_rate_pct']}%")

# ---------------------------
# 8) DISCOUNT IMPACT
print("\n=== 8) Discount Impact ===")
if discount_col not in df.columns:
    raise KeyError("Missing discount_percent column.")
df[discount_col] = pd.to_numeric(df[discount_col], errors='coerce').fillna(0)
avg_by_discount = df.groupby(discount_col)[total_value_col].mean().reset_index().sort_values(discount_col)
avg_by_discount.columns = [discount_col, "avg_total_value"]
display(avg_by_discount.head(40))
avg_by_discount.to_csv("avg_total_by_discount.csv", index=False)

plt.figure(figsize=(8,5))
plt.plot(avg_by_discount[discount_col], avg_by_discount["avg_total_value"], marker='o')
plt.xlabel("Discount Percent")
plt.ylabel("Average Total Value")
plt.title("Average Total Value by Discount Percent")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("discount_vs_avg_totalvalue.png", dpi=150)
print("Saved discount_vs_avg_totalvalue.png")

# Quick interpretation
if avg_by_discount.shape[0] >= 2:
    slope = (avg_by_discount["avg_total_value"].iloc[-1] - avg_by_discount["avg_total_value"].iloc[0]) / (avg_by_discount[discount_col].iloc[-1] - avg_by_discount[discount_col].iloc[0] + 1e-9)
    if slope < -1e-6:
        print("- Trend: higher discounts correlate with lower avg order value (possible discounting of low-ticket items).")
    elif slope > 1e-6:
        print("- Trend: higher discounts correlate with higher avg order value (discounts driving larger purchases).")
    else:
        print("- No obvious linear trend between discount and avg order value.")

# ---------------------------
# 9) FEEDBACK KEYWORDS
print("\n=== 9) Feedback Keywords ===")
if feedback_col not in df.columns:
    raise KeyError("Missing customer_feedback column.")
s = df[feedback_col].fillna("").astype(str).str.lower()
keywords = ["great", "poor", "delayed", "excellent"]
counts = {kw: int(s.str.contains(r"\b" + re.escape(kw) + r"\b", na=False).sum()) for kw in keywords}
print("Keyword counts:")
for k,v in counts.items():
    print(f"- {k}: {v}")
pd.Series(counts, name="count").to_csv("feedback_keyword_counts.csv")
print("Saved feedback_keyword_counts.csv")

# ---------------------------
# 10) MONTHLY TREND (2024 only)
print("\n=== 10) Monthly Trend for 2024 ===")
if date_col not in df.columns:
    raise KeyError("Missing order_date column.")
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
df["order_month"] = df[date_col].dt.to_period("M").astype(str)
df_2024 = df[df[date_col].dt.year == 2024].copy()
monthly_2024 = df_2024.groupby("order_month")[total_value_col].sum().reset_index().sort_values("order_month")
display(monthly_2024)
monthly_2024.to_csv("monthly_revenue_2024.csv", index=False)

plt.figure(figsize=(9,4))
plt.plot(monthly_2024["order_month"], monthly_2024[total_value_col], marker='o')
plt.xticks(rotation=45)
plt.xlabel("Month (2024)")
plt.ylabel("Total Revenue")
plt.title("Monthly Total Revenue (2024)")
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig("monthly_revenue_2024.png", dpi=150)
print("Saved monthly_revenue_2024.png")

# ---------------------------
# 11) CUSTOMER LOYALTY
print("\n=== 11) Customer Loyalty ===")
if cust_col not in df.columns:
    raise KeyError("Missing customer_id column.")
cust_stats = df.groupby(cust_col).agg(
    total_orders = pd.NamedAgg(column=order_col, aggfunc="count"),
    total_spent = pd.NamedAgg(column=total_value_col, aggfunc="sum"),
    avg_order_value = pd.NamedAgg(column=total_value_col, aggfunc=lambda x: round(x.mean(),2))
).reset_index().sort_values("total_spent", ascending=False)
top10_customers = cust_stats.head(10)
display(top10_customers)
top10_customers.to_csv("top10_customers_by_spending.csv", index=False)
print("Saved top10_customers_by_spending.csv")

# ---------------------------
# 12) SQL Query
print("\n=== 12) SQL (answer) ===")
sql_query = f"""
-- Top 3 product categories by average profit margin in the 'North' region, only delivered orders
SELECT
  {pc_col} AS product_category,
  AVG( profit_margin ) AS avg_profit_margin
FROM orders
WHERE region = 'North'
  AND delivery_status = 'Delivered'
GROUP BY {pc_col}
HAVING COUNT(*) >= 5    -- optional: minimum sample size
ORDER BY avg_profit_margin DESC
LIMIT 3;
"""
print(sql_query)

# ---------------------------
# List saved files for download
print("\nFiles created in Colab working directory:")
for fn in os.listdir('.'):
    if fn.endswith('.csv') or fn.endswith('.png'):
        print("- ", fn)
```
